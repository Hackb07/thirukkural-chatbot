```python

# Rename columns
df = df.rename(columns={
    "Chapter Name": "adhigaram_name",
    "Section Name": "paal",
    "Verse": "kural_tamil",
    "Translation": "meaning_english",
    "Explanation": "explanation"
})

# Add Kural ID
df["kural_id"] = range(1, len(df) + 1)

# Generate Adhigaram ID (10 kurals per adhigaram)
df["adhigaram_id"] = ((df["kural_id"] - 1) // 10) + 1

# Optional: derive Iyal (simple mapping)
def derive_iyal(paal):
    if "Aram" in paal:
        return "Paal-I"
    elif "Porul" in paal:
        return "Paal-II"
    else:
        return "Paal-III"

df["iyal"] = df["paal"].apply(derive_iyal)

# Reorder columns
df = df[
    [
        "kural_id",
        "paal",
        "iyal",
        "adhigaram_id",
        "adhigaram_name",
        "kural_tamil",
        "meaning_english",
        "explanation"
    ]
]

# Save final dataset
df.to_csv("thirukkural_chatbot_dataset.csv", index=False, encoding="utf-8-sig")

print("âœ… Chatbot-ready Thirukkural dataset created")
```

```python
import pandas as pd
import faiss
import pickle
from sentence_transformers import SentenceTransformer

# Load dataset
df = pd.read_csv("thirukkural_chatbot_dataset.csv")

# Combine text fields for better semantic understanding
df["combined_text"] = (
    df["kural_tamil"].astype(str) + " " +
    df["meaning_english"].astype(str) + " " +
    df["explanation"].astype(str) + " " +
    df["adhigaram_name"].astype(str)
)

# Load embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Generate embeddings
embeddings = model.encode(
    df["combined_text"].tolist(),
    show_progress_bar=True
)

# Create FAISS index
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)

# Save index and dataset
faiss.write_index(index, "thirukkural.faiss")
pickle.dump(df, open("thirukkural.pkl", "wb"))

print("âœ… FAISS index built successfully")
print(f"ğŸ”¢ Total vectors indexed: {index.ntotal}")
```

```python
#Build the Retrieval Engine
import faiss
import pickle
from sentence_transformers import SentenceTransformer

# Load FAISS index
index = faiss.read_index("thirukkural.faiss")

# Load dataset
df = pickle.load(open("thirukkural.pkl", "rb"))

# Load embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")

def retrieve_kurals(query, top_k=3):
    """
    Retrieves the most relevant Thirukkural(s) for a given query
    """
    # Encode user query
    query_embedding = model.encode([query])

    # Search FAISS
    distances, indices = index.search(query_embedding, top_k)

    results = []
    for idx in indices[0]:
        row = df.iloc[idx]
        results.append({
            "kural_id": row["kural_id"],
            "adhigaram": row["adhigaram_name"],
            "paal": row["paal"],
            "kural_tamil": row["kural_tamil"],
            "meaning_english": row["meaning_english"],
            "explanation": row["explanation"]
        })

    return results
```
